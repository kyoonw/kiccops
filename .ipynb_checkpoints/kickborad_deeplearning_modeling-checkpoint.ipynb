```python
# íŒ¨í‚¤ì§€ ì„¤ì¹˜ í›„ ë²„ì „í™•ì¸
!pip install -q ultralytics
import ultralytics
ultralytics.checks()
```

```python
# í† ì¹˜ ë²„ì „ í™•ì¸
import torch
torch.cuda.get_device_name(0)
torch.cuda.is_available()
print(torch.__version__)

# GPUê°€ ì‚¬ìš© ê°€ëŠ¥í•œì§€ í™•ì¸í•˜ê³ , ì‚¬ìš© ê°€ëŠ¥í•˜ë©´ CUDAë¥¼ ì‚¬ìš©í•˜ë„ë¡ ì„¤ì •
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")
```

```python
# YOLO ë¼ì´ë¸ŒëŸ¬ë¦¬ ê°€ì ¸ì˜¤
from ultralytics import YOLO

# 'yolov8n.pt' ëª¨ë¸ ì„ ì–¸í•˜ê¸° - ì‚¬ì „í•™ìŠµëœ YOLOv8n detection model ë¡œë“œí•˜ê¸°
model = YOLO('yolov8n.pt')
```

```python
model.train(data='/content/kickboard_data/kickboard_data.yaml', epochs=10, patience=8, batch=16, imgsz=416)

# í•™ìŠµì´ ëë‚œ í›„ ì „ì²´ ë°ì´í„°ì…‹ì— ëŒ€í•œ ì •í™•ë„ë¥¼ í‰ê°€í•©ë‹ˆë‹¤.
accuracy = model.evaluate(data='/content/kickboard_data/kickboard_data.yaml', batch_size=16, img_size=416)

# ì •í™•ë„ë¥¼ ì¶œë ¥í•©ë‹ˆë‹¤.
print(f"Accuracy: {accuracy}")
```
