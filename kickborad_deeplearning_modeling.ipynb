```python
# 패키지 설치 후 버전확인
!pip install -q ultralytics
import ultralytics
ultralytics.checks()
```

```python
# 토치 버전 확인
import torch
torch.cuda.get_device_name(0)
torch.cuda.is_available()
print(torch.__version__)

# GPU가 사용 가능한지 확인하고, 사용 가능하면 CUDA를 사용하도록 설정
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")
```

```python
# YOLO 라이브러리 가져오
from ultralytics import YOLO

# 'yolov8n.pt' 모델 선언하기 - 사전학습된 YOLOv8n detection model 로드하기
model = YOLO('yolov8n.pt')
```

```python
model.train(data='/content/kickboard_data/kickboard_data.yaml', epochs=10, patience=8, batch=16, imgsz=416)

# 학습이 끝난 후 전체 데이터셋에 대한 정확도를 평가합니다.
accuracy = model.evaluate(data='/content/kickboard_data/kickboard_data.yaml', batch_size=16, img_size=416)

# 정확도를 출력합니다.
print(f"Accuracy: {accuracy}")
```
